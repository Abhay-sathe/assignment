 ### Identify such ratings where review text is good, but rating is negative- so that the support team can point this to users.

```python
import pandas as pd
pd.set_option("max_colwidth",150)
```

```python
df = pd.read_csv(r"E:\\Interview purpose\\New folder\\chrome_reviews.csv")

def fun(text):
    words = ["good","nice","Good","Nice"]
    for word in words:
        if word in text:
            return True
    return False
```

```python
neg_review = df[(df['Star'] == 1) | (df['Star'] == 2)]
```

```python
neg_review = neg_review[neg_review['Text'].apply(fun)]
```

```python
def fun2(text):
    return not "not" in text
neg_review = neg_review[neg_review['Text'].apply(fun2)]
```

```python
neg_review['Developer Reply'] = ["Would you mind updating your rating according to your review!" for i in range(len(neg_review))]
```

### Ranking Data - Understanding the co-relation between keyword rankings with description or any other attribute.

```python
import pandas as pd
pd.set_option("max_colwidth",50)
import numpy as np
```

```python
df = pd.read_csv(r"E:\\Interview purpose\\New folder\\browser_rankings_data.csv")
df
```

```python
desc = df['Short Description'].unique()

df['short no'] = [np.nan for i in range(len(df))]

d = {desc[i] : i for i in range(len(desc))}

for i in range(len(df)):
    df['short no'][i] = d[df['Short Description'][i]]

desc = df['Long Description'].unique()

df['long no'] = [np.nan for i in range(len(df))]

d = {desc[i] : i for i in range(len(desc))}

for i in range(len(df)):
    df['long no'][i] = d[df['Long Description'][i]]
    
desc = df['App ID'].unique()

df['app id no'] = [np.nan for i in range(len(df))]

d = {desc[i] : i for i in range(len(desc))}

for i in range(len(df)):
    df['app id no'][i] = d[df['App ID'][i]]
    
desc = df['Keyword'].unique()

df['Keyword no'] = [np.nan for i in range(len(df))]

d = {desc[i] : i for i in range(len(desc))}

for i in range(len(df)):
    df['Keyword no'][i] = d[df['Keyword'][i]]
df
```
